# **Exploring and Visualizing a Simple Dataset**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Iris dataset from seaborn
iris = sns.load_dataset("iris")
```

```python
# Display dataset shape
print("Shape of dataset:", iris.shape)

# Display column names
print("Columns in dataset:", iris.columns)

# Display first 5 rows
print(iris.head())
```

```python
# Scatter plot of sepal_length vs sepal_width
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
# Histogram of petal_length
sns.histplot(data=iris, x="petal_length", kde=True)
plt.title("Distribution of Petal Length")
plt.show()
```

```python
# Box plot for sepal_length across species
sns.boxplot(data=iris, x="species", y="sepal_length")
plt.title("Box Plot of Sepal Length by Species")
plt.show()
```
